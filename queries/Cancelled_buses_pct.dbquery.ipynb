```sql
SELECT 
  prod_year,
  COUNT(*) AS total_transport,
  SUM(CASE WHEN canceled_dt IS NOT NULL THEN 1 ELSE 0 END) AS canceled_count,
  SUM(CASE WHEN canceled_dt IS NULL THEN 1 ELSE 0 END) AS not_canceled_count,
  ROUND(
    100.0 * SUM(CASE WHEN canceled_dt IS NOT NULL THEN 1 ELSE 0 END) / COUNT(*),
    2
  ) AS pct_canceled
FROM transport_lakehouse.gold.fact_public_vehicles
WHERE vehicle_type_key != 2 AND vehicle_type_key != 3 AND vehicle_type_key != 5 AND vehicle_type_key != 20
GROUP BY prod_year
ORDER BY prod_year;

```
