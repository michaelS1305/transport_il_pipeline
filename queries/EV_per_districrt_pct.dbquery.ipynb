```sql
 WITH Car_Count AS(
    SELECT district_nm, SUM(vehicle_count) AS vehicle_amount
    FROM transport_lakehouse.gold.fact_ev_counts_by_area
    GROUP BY district_nm
    ORDER BY vehicle_amount DESC
 )

 SELECT 
 district_nm,
 ROUND(100.0 * vehicle_amount / SUM(vehicle_amount) OVER (), 2) AS pct
 FROM Car_Count


```
