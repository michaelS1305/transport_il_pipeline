## INIT

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import StructType, StructField, StringType
```

## Reading from bronze layer

```python
df_bronze =spark.table("transport_lakehouse.bronze.electric_car_amount_per_region_raw")
df_bronze.display()
```

## Columns organization and new column aliases

```python
df_silver = df_bronze.select(
  F.col("mahoz_nm").cast("string").alias("district_nm"),
  F.col("nafa_nm").cast("string").alias("region_nm"),
   F.col("baalut").cast("string").alias("ownership"),
  F.col("car_num").cast("int").alias("vehicle_count")
)

df_silver.display()
```

## Writing to bronze layer

```python
(
    df_silver.write 
        .mode("overwrite") 
        .format("delta") 
        .option("overwriteSchema", "true")
        .saveAsTable("transport_lakehouse.silver.silver_vehicles_electric")

)

spark.table("transport_lakehouse.silver.silver_vehicles_electric").display()
```
