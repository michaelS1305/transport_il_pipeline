## Reading from bronze layer

```python
logger.info("Step 1/3: Reading data from Bronze layer...")
df_bronze =spark.table("transport_lakehouse.bronze.electric_car_amount_per_region_raw")
logger.info(f"Successfully read {df_bronze.count()} rows from Bronze.")
```

## Columns organization and new column aliases

```python
logger.info("Step 2/3: Starting column reorganization and aliasing...")
df_silver = df_bronze.select(
  F.col("mahoz_nm").cast("string").alias("district_nm"),
  F.col("nafa_nm").cast("string").alias("region_nm"),
   F.col("baalut").cast("string").alias("ownership"),
  F.col("car_num").cast("int").alias("vehicle_count")
)

logger.info("Step 2/5: Finished column reorganization.")
```

## Writing to silver layer

```python
target_table = "transport_lakehouse.silver.silver_vehicles_electric"
logger.info(f"Step 3/3: Writing final DataFrame to Silver table: {target_table}")
(
    df_silver.write 
        .mode("overwrite") 
        .format("delta") 
        .option("overwriteSchema", "true")
        .saveAsTable(target_table)

)

logger.info("--- Silver_Vehicles_electric Process Completed Successfully ---")
```
